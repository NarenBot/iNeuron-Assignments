```python
import pandas as pd
import numpy as np
pd.pandas.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/PramodShenoy/911-Calls/master/911.csv')
df.head()
```

```python
# 1) describe for cat and numerical column:

df.describe(include='all')
```

```python
# 2) lat type-cast it to int, perform groupby and apply count on it

df['lat'] = df['lat'].astype('int64')
count = df.groupby('lat').size().reset_index()
count
```

```python
# 3) how many people are from same zip location

grp = df.groupby('zip').size().reset_index()
grp.head()
```

```python
# 4) check timeStamp datatype and extract day in new col

ls = []
for i in range(len(df['timeStamp'])):
    ls.append(df['timeStamp'][i].split()[0].split('-'))
    
year = []
month = []
day = []
for i in range(len(df['timeStamp'])):
    year.append(ls[i][0])
    month.append(ls[i][1])
    day.append(ls[i][2])

    
df['day'] = day
df.head(2)
```

```python
# 5) filter out all the zip codes which address starts with 'A'

filter_addr = []
for i in range(len(df['addr'])):
    filter_addr.append(str(df['addr'][i]).startswith('A'))

filter_zip = []
for i in range(len(df['zip'])):
    if filter_addr[i]==True:
        filter_zip.append(df['zip'][i])
        
series = pd.Series(filter_zip)
series.unique()
```

```python
# 6) try to map timestamp with day like monday, tuesday....


ls = []
for i in range(len(df['timeStamp'])):
    ls.append(df['timeStamp'][i].split()[0].split('-'))
    
year = []
month = []
day = []
for i in range(len(df['timeStamp'])):
    year.append(ls[i][0])
    month.append(ls[i][1])
    day.append(ls[i][2])

import datetime
day_name = []
for i in range(len(df['timeStamp'])):
    res = datetime.date(int(year[i]),int(month[i]),int(day[i]))
    day_name.append(res.strftime('%A'))
    
df['dayName'] = day_name
df.head(2)
```

```python
# 7) from title extract service detail

service=[]
for i in range(len(df['title'])):
    if 'EMS' in df['title'][i]:
        service.append('Emergency Service')   
    elif 'Fire' in df['title'][i]:
        service.append('Fire Service')   
    elif 'Traffic' in df['title'][i]:
        service.append('Traffic Service')
        
df['service'] = service
df.head(2)
```
