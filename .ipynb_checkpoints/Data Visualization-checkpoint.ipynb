```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris = datasets.load_iris()
type(iris)
```

```python
x = iris['data']
y = iris['target']
y = pd.DataFrame(data= y, columns=['target'])
```

```python
pca = PCA(n_components=3)
pca_trans = pca.fit_transform(x)
```

```python
pca_df = pd.DataFrame(data= pca_trans, columns= ['PC1', 'PC2', 'PC3'])
pca_df.head()
```

```python
df = pd.concat([pca_df, y], axis=1)
df.head()
```

### Naming Conversion:

```python
for i in range(len(df['target'])):
    if df['target'][i] == 0:
        df['target'][i] = 'Setosa'
    elif df['target'][i] == 1:
        df['target'][i] = 'Versicolor'
    elif df['target'][i] == 2:
        df['target'][i] = 'Virginica'

df.head()
```

### Output Graph:

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d') ## Plot in 3D
ax.set_title('3-Component PCA', fontsize = 15)

targets = ['Setosa', 'Versicolor', 'Virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    class_index = df['target'] == target
    ax.scatter(df.loc[class_index, 'PC1']
               , df.loc[class_index, 'PC2']
               , df.loc[class_index, 'PC3']
               , c = color
               , s = 40)
ax.legend(targets)
ax.grid()
plt.show()
```
